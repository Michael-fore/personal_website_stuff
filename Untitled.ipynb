```python
def hex_to_rgb(color):
    color = color.lstrip('#')
    colors = []
    for i in (2,4,6):
       #print(color[i-2:i])
        colors.append(int(color[i-2:i], 16))
    return colors
```

```python
def handle_indices(indices, shape=(500,500), color=[255,255,255]):
    '''Takes in a _indices output and create a np.array of rgb values'''
    
    x,y = shape
    cc, rr = indices
    return_grid = [[[0,0,0] for i in range(0,y)] for i in range(0,x)] 
    
    for x,y in zip(rr,cc):
        return_grid[y][x] = color

    
    return np.array(return_grid).astype(np.uint8)
```

```python
def _indices_of_path(path, scale=1):
    """
    Retrieve pixel indices (integer values). 

    Parameters
    ----------

    path: SVG-like path formatted as JSON string
        The path is formatted like
        ['M', x0, y0],
        ['Q', xc1, yc1, xe1, ye1],
        ['Q', xc2, yc2, xe2, ye2],
        ...
        ['L', xn, yn]
        where (xc, yc) are for control points and (xe, ye) for end points.

    Notes
    -----

    I took a weight of 1 and it seems fine from visual inspection.
    """
    rr, cc = [], []
    for (Q1, Q2) in zip(path[:-2], path[1:-1]):
        # int(round()) is for Python 2 compatibility
        inds = draw.bezier_curve(int(round(Q1[-1] / scale)), 
                                 int(round(Q1[-2] / scale)), 
                                 int(round(Q2[2] / scale)), 
                                 int(round(Q2[1] / scale)), 
                                 int(round(Q2[4] / scale)), 
                                 int(round(Q2[3] / scale)), 1)
        rr += list(inds[0])
        cc += list(inds[1])
    return rr, cc
```

```python
def _indices_of_rect(obj, scale=1):
    
    scale_factor = obj['scaleX'] / scale
    props = [ scale_factor * obj['width'],
              scale_factor * obj['height'],
              scale_factor * obj['left'],
              scale_factor * obj['top']]
    
    r0,c0,r1,c1= np.array(props).astype(np.int8)
    
    rr, cc = draw.rectangle_perimeter((r0,c0),(r1,c1))
    
    return rr,cc
```

```python
# 'x1': -59.5, 'x2': 59.5, 'y1': -28.5, 'y2': 28.5

from skimage import draw


def _indices_of_line(obj, scale=1):
    
    ''' Takes a json object with type path'''
    
    length = np.sqrt(obj['width']**2 + obj['height']**2)
    
    scale_factor = obj['scaleX'] / scale
    
    props = [scale_factor * length,
              scale_factor * obj['width'],
              scale_factor * obj['height'],
              scale_factor * obj['left'],
              scale_factor * obj['top']]
    
    s,r0,c0,r1,c1= np.array(props).astype(np.int8)
    
    rr, cc = draw.line(r0,c0,r1,c1)
    
    return  rr, cc
```

```python
def canvas_json_handler(json, shape = (200,200), scale = 1):
    '''Take is the jsonified path from dash-canvas and converts each  path to a color map for further processing.
    return a list of numpy arrays with rgb values [[255,255,255]]'''

    color_maps = []
    for obj in json['objects']:
        if obj['type'] == 'line':
            rr, cc = _indices_of_line(obj, scale = scale)
            color =  hex_to_rgb(obj['stroke'])
            color_map = handle_indices((rr,cc), shape = shape, color = color)
            color_maps.append(color_map)
            
        elif obj['type'] == 'rect':
            rr, cc = _indices_of_rect(obj, scale = scale)
            color =  hex_to_rgb(obj['stroke'])
            color_map = handle_indices((rr,cc), shape = shape, color = color)
            color_maps.append(color_map)
            
        elif obj['type'] == 'path':
            rr, cc = _indices_of_path(obj['path'], scale = scale)
            color =  hex_to_rgb(obj['stroke'])
            color_map = handle_indices((rr,cc), shape = shape, color = color)
            color_maps.append(color_map)
            
        elif obj['type'] == 'image':
            pass
            
        else:
            raise TypeError (f"path['objects']['type'] needs to be 'line','rect',or'path' instead of {obj['type']}")
            
    return color_maps
```

```python
import numpy.ma as ma

def img_path_combine(img, paths):
    '''Img must be a PIL image, paths from canvas json handler'''
    
    if not isinstance(paths, list):
        paths = [paths]
    
    assert isinstance(img, Image.Image), 'Make sure img is a instance of PIL.Image.Image or Image.Image\
                                            if using from PIL import img'
    
    im = img.copy()
    
    images = map(lambda x: Image.fromarray(x), paths) 
    #print(len(list(images)))
    for i in images:
        mask = i.copy().convert('L').point(lambda x: 255 if x > 1 else 0)
        im.paste(i,mask=mask)
        
        break 
    return im
    
```
